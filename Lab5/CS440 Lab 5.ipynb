```python
import os
import pandas as pd

# Get each data file from Jim Weller's dataset
datapath = "/Users/hawk/Documents/DOCUMENTS/School/2014S/CMPSC440/Lab5/data"

def get_files(run):
    datafiles = {}
    runpath = (datapath + "/" + run)
    files = [f for f in os.listdir(runpath) if ('.csv' in f and "jobQueue.csv" not in f)]
    # loop through & load the data files -> Pandas data frames
    for filename in files:
        filepath = runpath + "/" + filename
        datafiles[run+"_"+filename] = pd.read_csv(filepath,
                                                  header=1,
                                                  skiprows=range(52,57))
    return pd.concat(datafiles) 
```

```python
#####################
# HELPER FUNCTIONS #
####################
def filenames (df):
    """Extracts filenames from a dataframe"""
    return df.reset_index().level_0.unique()

def list_waits (df):
    for name in sorted(filenames(df)):
        yield (name, df.xs((name), 0)['Wait'])
    
def list_responses (df):
    for name in sorted(filenames(df)):
        yield (name, df.xs((name), 0)['Response'])
    
def list_turnarounds (df):
    for name in sorted(filenames(df)):
        yield (name, df.xs((name), 0)['Turnaround'])
```

```python
##########################
# VIZ/ANALYSIS FUNCTIONS #
##########################
texstr = "Wait time &\t\t{wmean} &\t{wdev} &\t{wmin} &\t{wmax} \t\\\\\nResponse time &\t\t{rmean} &\t{rdev} &\t{rmin} &\t{rmax} \t\\\\\nTurnaround time &\t{tmean} &\t{tdev} &\t{tmin} &\t{tmax} \t\\\\"

def table (df):
    """Prints describe() tables for each algorithm"""
    for name in filenames(df):
        print(name + "\n-------------------------------------")
        print(df.xs(name, 0).describe())
        print("\n")
        
def textable(df):
    """Prints tables for each algorithm, ready to be pasted into a LaTeX Tabular environment"""
    for name in filenames(df):
        print(name +"\n\t\t\tmean\tstd.dev\t\tmin\tmax")
        print("\t\t\t----\t-------\t\t---\t---")
        print(texstr.format(
                            # wait time stats
                            wmean = np.mean(df.xs(name,0)["Wait"]),
                            wdev  = np.std(df.xs(name,0)["Wait"]),
                            wmin  = np.amin(df.xs(name,0)["Wait"]),
                            wmax  = np.amax(df.xs(name,0)["Wait"]),
                            # response time stats
                            rmean = np.mean(df.xs(name,0)["Response"]),
                            rdev  = np.std(df.xs(name,0)["Response"]),
                            rmin  = np.amin(df.xs(name,0)["Response"]),
                            rmax  = np.amax(df.xs(name,0)["Response"]),
                            # turnaround time stats
                            tmean = np.mean(df.xs(name,0)["Turnaround"]),
                            tdev  = np.std(df.xs(name,0)["Turnaround"]),
                            tmin  = np.amin(df.xs(name,0)["Turnaround"]),
                            tmax  = np.amax(df.xs(name,0)["Turnaround"]),
                            )
              )
        print("\n")
        
```

```python
def merge_rands(measurement):
    result = {}
    if measurement == "waits":
        rands = [dict(list_waits(rand1)),
                dict(list_waits(rand2)),
                dict(list_waits(rand3)),
                dict(list_waits(rand4)),
                dict(list_waits(rand5))]
    elif measurement == "responses":
        rands = [dict(list_responses(rand1)),
                dict(list_responses(rand2)),
                dict(list_responses(rand3)),
                dict(list_responses(rand4)),
                dict(list_responses(rand5))]
    elif measurement == "turnarounds":
        rands = [dict(list_turnarounds(rand1)),
                dict(list_turnarounds(rand2)),
                dict(list_turnarounds(rand3)),
                dict(list_turnarounds(rand4)),
                dict(list_turnarounds(rand5))]
    else: rands = []
        
    sjf_pres = []
    sjfs = []
    rrs = []
    rr_pris = []
    fcfss = []
    priorities = []
    priorities_pre = []
        
    for run in rands:
        for key, value in run.items():
            if "sjf-preemptive" in key:
                sjf_pres.append(value)
            elif "sjf" in key:
                sjfs.append(value)
            elif "rr-pri" in key:
                rr_pris.append(value)
            elif "rr" in key:
                rrs.append(value)
            elif "priority-preemptive" in key:
                priorities_pre.append(value)
            elif "priority" in key:
                priorities.append(value)
            elif "fcfs" in key:
                fcfss.append(value)
                
    return [pd.concat(priorities),
            pd.concat(priorities_pre),
            pd.concat(rrs),
            pd.concat(rr_pris),
            pd.concat(sjfs),
            pd.concat(sjf_pres),
            pd.concat(fcfss)]
```

```python
plt.figure(4, figsize=(9,6))
plt.title("Wait time ~ scheduling algorithm")

sns.boxplot(merge_rands("waits"),
            color = 'BuGn',
            notch = True,
            names = names)

sns.axlabel("", "Wait time (CPU cycles)")
sns.despine()

plt.savefig(figpath + 'rand_whisker_wait.pdf')

plt.figure(5, figsize=(9,6))
plt.title("Response time ~ scheduling algorithm")

sns.boxplot(merge_rands("responses"), 
            color = 'BuGn', 
            notch = True,
            names = names)

sns.axlabel("", "Response time (CPU cycles)")
sns.despine()

plt.savefig(figpath + 'rand_whisker_response.pdf')

plt.figure(6, figsize=(9,6))
plt.title("Turnaround time ~ scheduling algorithm")

sns.boxplot(merge_rands("turnarounds"), 
            color = 'BuGn', 
            notch = True,
            names = names)
    
sns.axlabel("", "Turnaround time (CPU cycles)")
sns.despine()

plt.savefig(figpath + 'rand_whisker_turnaround.pdf')
```
